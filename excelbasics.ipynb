 """Relative cell referncing can be defined as fully flexible, it does not contain $ so that it can apply same calculation within multiple rows and columns. eg .     ###Absolute Cell Refrence
"""Absolute cell refrencing can be defined as fully fixed, it contains $ so that it can lock constant values.for eg. if we want to write tables(table of 2,3, etc) for that we can use this reference."""
